```python
df_train["relative_mertebe"] = df_train["Mərtəbə"].apply(lambda x: int(x.split(" / ")[0]) / int(x.split(" / ")[1]))
df_test["relative_mertebe"] = df_test["Mərtəbə"].apply(lambda x: int(x.split(" / ")[0]) / int(x.split(" / ")[1]))
```

```python
#remove outliers

column='price'
q_low = df_train[column].quantile(0.01)
q_hi  = df_train[column].quantile(0.99)
df_train = df_train[(df_train[column] < q_hi) & (df_train[column] > q_low)]
df_train.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X = df_train[["Sahə","Otaq sayı","exact_mertebe","relative_mertebe","baki","sumqayit","xirdalan","longitude","latitude", "merkeze_yaxinliq", "sebail", "nesimi", 'isiq', 'qaz', 'su', 'parkinq', 'white', 'kompleks']].values.reshape(-1,18)
y = df_train["price"].values

# scaler = StandardScaler()
# scaled_X = scaler.fit_transform(X)
# X = scaled_X
X_test_sub = df_test[["Sahə","Otaq sayı","exact_mertebe","relative_mertebe","baki","sumqayit","xirdalan","longitude","latitude", "merkeze_yaxinliq", "sebail", "nesimi", 'isiq', 'qaz', 'su', 'parkinq', 'white', 'kompleks']].values.reshape(-1,18)
# scaled_X_sub = scaler.transform(X_test_sub)
# X_test_sub = scaled_X_sub


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

model_xgb = XGBRegressor(colsample_bytree =1,learning_rate = 0.1,max_depth=18,min_child_weight= 1,n_estimators=100,subsample=0.75)
model_xgb_sub = XGBRegressor(colsample_bytree =0.75,learning_rate = 0.05,max_depth=18,min_child_weight= 1,n_estimators=100,subsample=0.75)


model_xgb.fit(X_train, y_train,
              verbose=True)
model_xgb_sub.fit(X, y,verbose=True)

y_pred_xgb = model_xgb.predict(X_test)
y_pred_xgb_sub = model_xgb_sub.predict(X_test_sub)

rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print(f'RMSE_xgb: {int(rmse_xgb)}')
y_pred_xgb
```

```python
df_sub=pd.DataFrame()
df_sub['_id']=df_test._id
df_sub['price']=pd.DataFrame(y_pred_xgb_sub)

df_sub.to_csv("res_xgb.csv",index=False)
```

```python
from catboost import CatBoost
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score, precision_score, recall_score

scorers = {
        'precision_score': make_scorer(precision_score),
        'accuracy_score': make_scorer(accuracy_score)
        }



model_cb = CatBoost() 
    
parameters = {'depth' : [8, 10, 12],
                'learning_rate' : [0.01,0.1, 0.5, 1],
                'iterations' : [30,40,50,60,70,80]}
    
Grid_CB = GridSearchCV(estimator=model_cb, param_grid = parameters, cv = 3, scoring=scorers, refit="precision_score")
Grid_CB.fit(X_train, y_train)

Grid_CB.best_params_
```
