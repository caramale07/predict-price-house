```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df=pd.read_csv('binaaz_train.csv')
```

```python
test_df=pd.read_csv('binaaz_test.csv')
```

```python
sample_df=pd.read_csv('binaaz_sample.csv')
```

```python
train_df['seher'].unique()
```

```python
def seher_to_int(x):
    if x=='baki':
        return 0
    elif x=='xirdalan':
        return 1
    else:
        return 2
train_df['seher'] = train_df['seher'].apply(lambda x: seher_to_int(x))      
```

```python
def to_float(x):
    return float(str(x).replace(' ', '').replace('m²', '').replace('м²', ''))

train_df['Sahə'] = train_df['Sahə'].apply(lambda x: to_float(x))                 
```

```python
def to_binary(x):
    if x=='var':
        return 1
    else:
        return 0

train_df['Kupça'] = train_df['Kupça'].apply(lambda x: to_binary(x))   
train_df['İpoteka'] = train_df['İpoteka'].apply(lambda x: to_binary(x))   
```

```python
def floor_to_float(x):
    n=x.split()
    return float(int(n[0])/int(n[-1]))

train_df['Mərtəbə'] = train_df['Mərtəbə'].apply(lambda x: floor_to_float(x))   
```

```python
def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mean+(n_std*sd))]
        
    return df

remove_outliers(train_df, ['Sahə','Kupça','Mərtəbə', 'Otaq sayı', 'price'], 1)
```

```python
test_df['Sahə'] = test_df['Sahə'].apply(lambda x: to_float(x))
test_df['Kupça'] = test_df['Kupça'].apply(lambda x: to_binary(x))   
test_df['Mərtəbə'] = test_df['Mərtəbə'].apply(lambda x: floor_to_float(x))   
test_df['İpoteka'] = test_df['İpoteka'].apply(lambda x: to_binary(x))   
test_df['seher'] = test_df['seher'].apply(lambda x: seher_to_int(x))   
```

```python
X_train = train_df[['Otaq sayı', 'Sahə', 'Kupça', 'Mərtəbə', 'latitude', 'longitude', 'seher']]
y_train = train_df['price']

X_test=test_df[['Otaq sayı', 'Sahə', 'Kupça', 'Mərtəbə','latitude', 'longitude', 'seher']]
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor()
# Train the model on training data
rf.fit(X_train, y_train);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
```

```python
sample_df['price']=predictions
sample_df.to_csv('res_rfr.csv', index=False)
```
